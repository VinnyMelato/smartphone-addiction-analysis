```python
df = pd.read_csv("Smartphone_Addiction_Analysis.csv", sep=';')
```

```python
df['daily_screen_time_hours'] = pd.to_numeric(df['daily_screen_time_hours'], errors='coerce')
df['weekend_screen_time'] = pd.to_numeric(df['weekend_screen_time'], errors='coerce')

df['weekly_total_usage'] = df['daily_screen_time_hours'] + df['weekend_screen_time']

print(df.info())
```

```python
print(df.groupby('gender')['weekly_total_usage'].mean())
```

```python
print(df['weekly_total_usage'].describe())
```

```python
import matplotlib.pyplot as plt

plt.hist(df['weekly_total_usage'], bins=20)
plt.title("Distribution of Weekly Usage")
plt.xlabel("Hours")
plt.ylabel("Number of Users")
plt.show()
```

```python
df.groupby('gender')['weekly_total_usage'].mean().plot(kind='bar')
plt.title("Average Weekly Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Hours")
plt.show()
```

```python
def categorize(x):
    if x < 20:
        return "Low"
    elif x <= 40:
        return "Moderate"
    else:
        return "High"

df['usage_category'] = df['weekly_total_usage'].apply(categorize)

print(df['usage_category'].value_counts())
```
